```python
import pandas as pd
```

```python
df = pd.read_json('../data/2024-12-2-yc-companies.jl', lines=True)
df.head()
```

```python
df_fall = pd.read_json('../data/f24-output.jl', lines=True)
df_fall.head()
```

```python
print(f"DF lenght: {df_fall.shape[0]}")
df_fall.drop_duplicates(subset=['company_name', 'short_description'], inplace=True)
print("DF lenght:" , df_fall.shape[0])
```

```python
# Simple concatenation
combined_df = pd.concat([df, df_fall], ignore_index=True)
```

```python
print(f"DF Length: ", combined_df.shape[0])
combined_df.drop_duplicates(subset = ['company_name', 'short_description'], inplace=True)
print(f"DF Length: ", combined_df.shape[0])
```

```python
print(f"# of fall companies {len(combined_df[combined_df['batch'] == 'W24'])}")
```

```python
combined_df.to_csv('2024-12-3-all-yc-companies.csv', index=False)
```
