```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/2024-12-3-all-yc-companies.csv')
df.head()
```

```python
print(f"Number of fall companies {len(df[df["batch"] == "F24"])}")
```

```python
batch_order = {
    'W25': 0,    # Winter 2025
    'F24': 1,    # Fall 2024
    'S24': 2,    # Summer 2024
    'W24': 3,    # Winter 2024
    'S23': 4,    # Summer 2023
    'W23': 5,    # Winter 2023
    'S22': 6,    # Summer 2022
    'W22': 7,    # Winter 2022
    'S21': 8,    # Summer 2021
    'W21': 9,    # Winter 2021
    'S20': 10,   # Summer 2020
    'W20': 11,   # Winter 2020
    'S19': 12,   # Summer 2019
    'W19': 13,   # Winter 2019
    'S18': 14,   # Summer 2018
    'W18': 15,   # Winter 2018
    'S17': 16,   # Summer 2017
    'W17': 17,   # Winter 2017
    'IK12': 18,  # IK12
    'S16': 19,   # Summer 2016
    'W16': 20,   # Winter 2016
    'S15': 21,   # Summer 2015
    'W15': 22,   # Winter 2015
    'S14': 23,   # Summer 2014
    'W14': 24,   # Winter 2014
    'S13': 25,   # Summer 2013
    'W13': 26,   # Winter 2013
    'S12': 27,   # Summer 2012
    'W12': 28,   # Winter 2012
    'S11': 29,   # Summer 2011
    'W11': 30,   # Winter 2011
    'S10': 31,   # Summer 2010
    'W10': 32,   # Winter 2010
    'S09': 33,   # Summer 2009
    'W09': 34,   # Winter 2009
    'S08': 35,   # Summer 2008
    'W08': 36,   # Winter 2008
    'S07': 37,   # Summer 2007
    'W07': 38,   # Winter 2007
    'S06': 39,   # Summer 2006
    'W06': 40,   # Winter 2006
    'S05': 41    # Summer 2005
}
```

```python
# Sort the DataFrame
sorted_df = df.sort_values(
    by='batch',
    key=lambda x: x.map(batch_order),
    ascending=True
).reset_index(drop=True)

# Verify the order
print("Batch distribution after sorting:")
print(sorted_df['batch'].value_counts())
```
